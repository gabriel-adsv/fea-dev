```python
# Agora est√° por conta pr√≥pria e pronto para brilhar! üåü

### üîç **Pergunta Norteadora** ###
# Quais fatores demogr√°ficos, acad√™micos e comportamentais mais influenciam a classifica√ß√£o e a m√©dia de notas dos estudantes?

# üõ†Ô∏è **Dicas para come√ßar:**
# - D√™ uma boa olhada no dataset.
# - Explore as rela√ß√µes entre a vari√°vel `Classificacao_Notas` e outros fatores, como g√™nero, horas de estudo, faltas, etc.
# - Use gr√°ficos e tabelas para visualizar os padr√µes encontrados.
# - Escreva suas reflex√µes e hip√≥teses sobre o que voc√™ encontrou.
# - E o mais importante: divirta-se no processo de an√°lise! üòÑ

# N√£o h√° limite para o n√∫mero de c√©lulas ou an√°lises. O espa√ßo √© todo seu! üöÄ

# **SEU C√ìDIGO COME√áA AQUI** üëá
import pandas as pd

### visao geral do dataset ###
traducao_colunas = {
  "StudentID": "ID_Aluno",
  "Age": "Idade",
  "Gender": "Genero",
  "Ethnicity": "Etnia",
  "ParentalEducation": "Escolaridade_Pais",
  "StudyTimeWeekly": "Horas_Estudo_Semanais",
  "Absences": "Faltas",
  "Tutoring": "Aulas_Particulares",
  "ParentalSupport": "Apoio_Pais",
  "Extracurricular": "Atividades_Extracurriculares",
  "Sports": "Esportes",
  "Music": "Musica",
  "Volunteering": "Voluntariado",
  "GPA": "Media_Notas",
  "GradeClass": "Classificacao_Notas"
}


dt = pd.read_csv('../data/dados_estudantes.csv', sep = ',')
dt = dt.rename(traducao_colunas, axis = 1) # traducao do nome das colunas

dt.info()
dt

```

```python
pd.options.display.float_format = '{:.1f}'.format # definicao de visualizacao para numeros com uma casa decimal

dt.isnull().sum() # contagem de valores nulos
dt.duplicated().sum() # contagem de linhas duplicadas

# substituicao de valores das colunas
dt["Genero"] = dt.apply(
  lambda row: 'masculino' if row["Genero"] == 0
  else 'feminino',
  axis = 1
)

dt["Etnia"] = dt.apply(
  lambda row: 'caucasiano' if row["Etnia"] == 0
  else 'afrodescendente' if row["Etnia"] == 1
  else 'asiatico' if row["Etnia"] == 2 
  else 'outro',
  axis = 1
)

dt["Escolaridade_Pais"] = dt.apply(
  lambda row: 'nenhuma' if row["Escolaridade_Pais"] == 0
  else 'ensino_medio' if row["Escolaridade_Pais"] == 1
  else 'graduacao_incompleta' if row["Escolaridade_Pais"] == 2
  else 'graduacao_completa' if row["Escolaridade_Pais"] == 3
  else 'pos_graduacao',
  axis = 1
)

dt["Aulas_Particulares"] = dt.apply(
  lambda row: 'nao' if row["Aulas_Particulares"] == 0
  else 'sim',
  axis = 1
)

dt["Apoio_Pais"] = dt.apply(
  lambda row: 'nenhum' if row["Apoio_Pais"] == 0
  else 'baixo' if row["Apoio_Pais"] == 1
  else 'moderado' if row["Apoio_Pais"] == 2
  else 'alto' if row["Apoio_Pais"] == 3
  else 'muito_alto',
  axis = 1
)

dt["Atividades_Extracurriculares"] = dt.apply(
  lambda row: 'nao' if row["Atividades_Extracurriculares"] == 0
  else 'sim',
  axis = 1
)

dt["Esportes"] = dt.apply(
  lambda row: 'nao' if row["Esportes"] == 0
  else 'sim',
  axis = 1
)

dt["Musica"] = dt.apply(
  lambda row: 'nao' if row["Musica"] == 0
  else 'sim',
  axis = 1
)

dt["Voluntariado"] = dt.apply(
  lambda row: 'nao' if row["Voluntariado"] == 0
  else 'sim',
  axis = 1
)

dt["Classificacao_Notas"] = dt.apply(
  lambda row: 'A' if row["Classificacao_Notas"] == 0
  else 'B' if row["Classificacao_Notas"] == 1
  else 'C' if row["Classificacao_Notas"] == 2
  else 'D' if row["Classificacao_Notas"] == 3
  else 'F',
  axis = 1
)

dt
```

```python
### organizacao de colunas do dataset ###

# adicao das colunas 'esportes', 'musica' e 'voluntariado' em 'atividades extracurriculares'
dt["Atividades_Extracurriculares"] = dt.apply(
  lambda x: 'sim' if x["Atividades_Extracurriculares"] == 'sim' 
  or 'sim' in x[["Esportes", "Musica", "Voluntariado"]].values 
  else 'nao',
  axis = 1
)

# remocao das colunas 'esportes', 'musica' e 'voluntariado'
dt = dt.drop(columns=["Esportes", "Musica", "Voluntariado"])

# renomear coluna 'horas_estudo_semanais' para 'horas_estudo'
dt = dt.rename(columns={'Horas_Estudo_Semanais': 'Horas_Estudo'})

dt
```

```python
### identificacao de correlacoes secundarias ###

# pergunta norteadora: quais fatores demogr√°ficos, acad√™micos e comportamentais mais influenciam a classifica√ß√£o e a m√©dia de notas dos estudantes?

## dados demograficos
# - idade
# - genero
# - etnia
# - escolaridade dos pais

## dados academicos
# - aulas particulares
# - atividades extracurriculares

## dados comportamentais
# - apoio dos pais
# - horas de estudos semanais
# - faltas

import seaborn as sns

sns.set_context(font_scale = 1.2) # definicao do tamanho padrao da fonte

dt.describe() 
```

```python
import matplotlib.pyplot as plt

# quantidade de estudantes participantes da pesquisa por genero, idade e etnia
contagem_genero = dt['Genero'].value_counts()
contagem_idade = dt['Idade'].value_counts()
contagem_etnia = dt['Etnia'].value_counts()

plt.figure(figsize=(15, 5))

# grafico 1
plt.subplot(1, 3, 1)
plt.title("Estudantes por G√™nero")
grafico = sns.barplot(hue = contagem_genero.index, y = contagem_genero.values, palette = 'Set1')
for valor in grafico.containers:
  grafico.bar_label(valor, fmt='%d', label_type='edge')
plt.xlabel('G√™nero')
plt.ylabel('Contagem')
plt.tight_layout()

# grafico 2
plt.subplot(1, 3, 2)
plt.title("Estudantes por Idade")
grafico = sns.barplot(hue = contagem_idade.index, y = contagem_idade.values, palette = 'Set2')
for valor in grafico.containers:
  grafico.bar_label(valor, fmt='%d', label_type='edge')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.tight_layout()

# grafico 3
plt.subplot(1, 3, 3)
plt.title("Estudantes por Etnia")
plt.pie(contagem_etnia, labels = contagem_etnia.index, autopct="%1.1f%%", colors = sns.color_palette('Set3'))
plt.tight_layout()

plt.show()
```

```python
# proporcao escolaridade pais x etnia dos alunos
dados_caucasiano = dt[dt['Etnia'] == 'caucasiano']
dados_afrodescendente = dt[dt['Etnia'] == 'afrodescendente']
dados_asiatico = dt[dt['Etnia'] == 'asiatico']
dados_outro = dt[dt['Etnia'] == 'outro']

contagem_escolaridade_caucasiano = dados_caucasiano['Escolaridade_Pais'].value_counts()
contagem_escolaridade_afro = dados_afrodescendente['Escolaridade_Pais'].value_counts()
contagem_escolaridade_asiatico = dados_asiatico['Escolaridade_Pais'].value_counts()
contagem_escolaridade_outro = dados_outro['Escolaridade_Pais'].value_counts()

plt.figure(figsize=(15,15))

# estudantes caucasianos
plt.subplot(2,2,1)
plt.title("Propor√ß√£o da Escolaridade dos Pais de estudantes Caucasianos")
plt.pie(contagem_escolaridade_caucasiano, labels = contagem_escolaridade_caucasiano.index, autopct="%1.1f%%", colors = sns.color_palette('Pastel1'))

# estudantes afrodescendentes
plt.subplot(2,2,2)
plt.title("Propor√ß√£o da Escolaridade dos Pais de estudantes Afrodescendentes")
plt.pie(contagem_escolaridade_afro, labels = contagem_escolaridade_afro.index, autopct="%1.1f%%", colors = sns.color_palette('Pastel1'))

# estudantes asiaticos
plt.subplot(2,2,3)
plt.title("Propor√ß√£o da Escolaridade dos Pais de estudantes Asi√°ticos")
plt.pie(contagem_escolaridade_asiatico, labels = contagem_escolaridade_asiatico.index, autopct="%1.1f%%", colors = sns.color_palette('Pastel1'))

# estudantes outros
plt.subplot(2,2,4)
plt.title("Propor√ß√£o da Escolaridade dos Pais de estudantes de Outra etnia")
plt.pie(contagem_escolaridade_outro, labels = contagem_escolaridade_outro.index, autopct="%1.1f%%", colors = sns.color_palette('Pastel1'))

plt.tight_layout()
plt.show()
```

```python
# genero x atividades curriculares
genero_por_atividades = dt.groupby(['Genero','Atividades_Extracurriculares']).size().reset_index(name = 'Contagem')

plt.figure(figsize=(17,6))

plt.subplot(1,2,2)
plt.title("Rela√ß√£o entre G√™nero e realiza√ß√£o de Atividades Curriculares")
grafico = sns.barplot(data = genero_por_atividades, x = 'Genero', y = 'Contagem', hue = 'Atividades_Extracurriculares', palette='flare')
for valor in grafico.containers:
  grafico.bar_label(valor, fmt='%d', label_type='edge')
plt.xlabel("G√™nero")
plt.ylabel("Atividades Extracurriculares")
plt.tight_layout()

# atividades extracurriculares x quantidade de faltas
atividades_por_faltas = dt.groupby('Atividades_Extracurriculares')['Faltas'].sum().reset_index(name='Total_Faltas')

plt.subplot(1,2,1)
plt.title("Rela√ß√£o entre realiza√ß√£o de Atividades Extracurriculares e total de Faltas de estudantes")
grafico = sns.barplot(data = atividades_por_faltas, x = 'Atividades_Extracurriculares', y = 'Total_Faltas', hue = 'Atividades_Extracurriculares', palette='flare')
for valor in grafico.containers:
  grafico.bar_label(valor, fmt='%d', label_type='edge')
plt.xlabel("Atividades Extracurriculares")
plt.ylabel("Quantidade de Faltas Totais")
plt.tight_layout()

plt.show()
```

```python
# apoio dos pais x horas de estudo
apoio_por_horas_estudo = dt.groupby('Apoio_Pais')['Horas_Estudo'].mean().reset_index(name = 'Media_Horas')

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
plt.title("Rela√ß√£o entre Apoio do Pais e m√©dia de Horas de Estudos Semanais dos estudantes")
grafico = sns.barplot(data = apoio_por_horas_estudo, x = 'Apoio_Pais', y = 'Media_Horas', hue = 'Apoio_Pais', palette='cool')
for valor in grafico.containers:
  grafico.bar_label(valor, fmt='%.1f', label_type='edge')
plt.xlabel("Apoio dos Pais")
plt.ylabel("M√©dia de Horas de Estudo Semanais")
plt.tight_layout()

# idade x horas de estudo
idade_por_horas_estudo = dt.groupby('Idade')['Horas_Estudo'].mean().reset_index(name = 'Media_Horas')

plt.subplot(1,2,2)
plt.title("Rela√ß√£o entre Idade e m√©dia de Horas de Estudos Semanais dos estudantes")
grafico = sns.barplot(data = idade_por_horas_estudo, x = 'Idade', y = 'Media_Horas', hue = 'Idade', palette='cool')
for valor in grafico.containers:
  grafico.bar_label(valor, fmt='%.1f', label_type='edge')
plt.xlabel("Idade")
plt.ylabel("M√©dia de Horas de Estudo Semanais")
plt.tight_layout()

plt.show()
```
