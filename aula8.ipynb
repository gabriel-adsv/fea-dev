## 3.1. Gráfico de Linha: Evolução de Vendas do Semestre

<code>**plt.plot(x, y**, color = ' ', linestyle = ' ', linewidth = ' ')</code>

Imagine que você tem dados de vendas mensais de um semestre. Um gráfico de linha pode ajudar a visualizar como as vendas evoluíram ao longo do tempo.

```python
import matplotlib.pyplot as plt

# Dados: Vendas mensais de um ano
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'] # eixo x
vendas = [2000, 2500, 2700, 2200, 2600, 3000] # eixo y

# Criando o gráfico de linha = plt.plot
plt.plot(meses, vendas, marker='o', color='b', linestyle='-', linewidth=2)

# marker: estilo de pontos no gráfico:
# 'o' - Círculo (padrão)
# 's' - Quadrado
# '^' - Triângulo para cima

# color: é a cor do gráfico:
#'r' para vermelho (red)
#'g' para verde (green)
#'b' para azul (blue)

# linestyle: Define o estilo da linha:
# '-' representa uma linha contínua.
# '--' para linha tracejada
# ':' para linha pontilhada
# '-.' para linha pontilhada e tracejada

# linewidth: define a espessura da linha. Geralmente o padrão é 1

# Títulos e rótulos
# adicionar título
plt.title('Vendas ao Longo do Ano')

# adicionar legenda aos eixos
plt.xlabel('Meses')
plt.ylabel('Vendas (em R$)')

# Mostrar gráfico
plt.show()
```

## 3.2. Gráfico de Dispersão: Análise de Altura e Peso

<code>**plt.scatter(x, y**, color = ' ', market = ' ')</code>

Suponha que você tem dados sobre a altura e peso de um grupo de pessoas. O gráfico de dispersão ajuda a visualizar a relação entre as duas variáveis.

```python
import matplotlib.pyplot as plt

# Dados: Altura (em cm) e Peso (em kg)
altura = [150, 160, 170, 180, 190, 200]
peso = [50, 60, 70, 80, 90, 100]

# Criando o gráfico de dispersão = plt.scatter
plt.scatter(altura, peso, color='red', marker='s')

# Relembrando:
# marker: estilo de pontos no gráfico (s = quadrado)
# color: é a cor do gráfico ('red' ou 'r')

# Títulos e rótulos
# define o título
plt.title('Relação entre Altura e Peso')

# define a legenda
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

# Mostrar gráfico
plt.show()
```

## 3.3 Gráfico de Barras: Comparação de Despesas por Categoria

<code>**plt.bar(x, y**, color = ' ')</code>

Você tem o orçamento de despesas mensais, e quer comparar as despesas por categoria. O gráfico de barras é ideal para isso.

```python
import matplotlib.pyplot as plt

# Dados: Categorias de despesas e seus valores
categorias = ['Alimentação', 'Transporte', 'Saúde', 'Educação', 'Lazer']
valores = [1200, 500, 300, 800, 200]

# Criando o gráfico de barras = plt.bar()
plt.bar(categorias, valores, color='g')

# color: g = green

# Títulos e rótulos
plt.title('Despesas Mensais por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Valor (em R$)')

# Mostrar gráfico
plt.show()
```

## 3.4. Histograma: Distribuição de Idades

</code>**plt.hist(x, bins = ' '**, color = ' ', edgecolor = ' ')<code>

Imaginemos que temos os dados de idades dos funcionários de uma empresa. Vamos usar esses dados para criar um histograma.

```python
import matplotlib.pyplot as plt

# Dados (idades dos funcionários de uma empresa)
idades = [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]

# Criando o histograma
plt.hist(idades, bins=5, color='lightgreen', edgecolor='black')

# bins: Definimos 5 faixas etárias para agrupar os dados.
# A faixa etária será distribuída de maneira que possamos ver quantos funcionários estão em cada faixa de idade.

# color: Define a cor do preenchimento
# edgecolor: Define a cor da borda das barras

# Adicionando título e rótulos aos eixos
plt.title('Distribuição das Idades dos Funcionários')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Funcionários')

# Exibindo o gráfico
plt.show()
```

## 3.5. Gráfico de Pizza: Participação de Mercado por Empresa

<code>**plt.pie(x, labels = nomes, autopct = '%1.1f%%', startangle = 90)**</code>

Vamos dizer que você tem a participação de mercado de 4 empresas no setor de tecnologia. Um gráfico de pizza pode ser útil para visualizar essa divisão.

```python
import matplotlib.pyplot as plt

# Dados: Participação de mercado por empresa
empresas = ['Apple', 'Samsung', 'Google', 'Microsoft']
participacao = [40, 30, 20, 10]

# Criando o gráfico de pizza = plt.pie
# note que as cores são definidas na ordem que estão definidas as empresas
plt.pie(participacao, labels=empresas, autopct='%1.1f%%', startangle=90, colors=['green','red','yellow','purple'])

# labels define os rótulos (nomes) que serão exibidos para cada "fatias" do gráfico de pizza.
# autopct ='%1.1f%%' é um formato de string que indica que as % serão exibidas com 1 casa decimal, seguidas pelo símbolo de %.
# startangle determina o ângulo de rotação inicial para o gráfico de pizza.

# Título
plt.title('Participação de Mercado')

# Mostrar gráfico
plt.show()
```

## Gráfico de linha

**Marker = 'o'**

Define o estilo de pontos no gráfico:
- 'o' - Círculo (padrão)
- 's' - Quadrado
- '^' - Triângulo para cima
- 'v' - Triângulo para baixo
- '>' - Triângulo para a direita
- '<' - Triângulo para a esquerda
- 'D' - Losango
- 'P' - Pentágono
- 'X' - Cruz
- '+' - Mais
- '.' - Ponto pequeno
- ',' - Ponto muito pequeno

**linestyle = '-'**

Define o estilo da linha.

- '-' representa uma linha contínua.

Outros estilos de linha incluem:
- '--' para linha tracejada
- ':' para linha pontilhada
- '-.' para linha pontilhada e tracejada

**linewidth = 2**

Define a espessura da linha.
- O valor padrão geralmente é 1, mas você pode aumentar ou diminuir o valor para tornar a linha mais grossa ou mais fina.

- linewidth = 2 significa que a linha terá uma espessura de 2 unidades.

```python
#EXEMPLO: 
# Dados
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

# Criando o gráfico com parâmetros personalizados
plt.plot(x, y, color='b', linestyle='-', linewidth=2, label='Linha Azul')

# Títulos e rótulos
plt.title('Exemplo de Gráfico com Estilo de Linha Personalizado')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Exibindo legenda
plt.legend()

# Mostrar gráfico
plt.show()
```

## Gráfico de pizza

**labels = empresas**

labels define os rótulos (nomes) que serão exibidos para cada "fatias" do gráfico de pizza.

No caso de um gráfico de pizza representando a participação de diferentes empresas no mercado, esse parâmetro pode ser uma lista de strings com os nomes das empresas.

Cada fatia da pizza será associada a um nome (ou rótulo), facilitando a identificação da participação de cada empresa.

**autopct = '%1.1f%%'**

autopct define o formato de exibição da porcentagem dentro de cada fatia do gráfico de pizza.

O '%1.1f%%' é um formato de string que indica que as porcentagens serão exibidas com 1 casa decimal, seguidas pelo símbolo de porcentagem.

Se você tiver 3 empresas e uma delas representar 30% do total, será exibido algo como 30.0% no gráfico.

autopct='%1.1f%%'  # Exibe a porcentagem com 1 casa decimal
autopct='%1.2f%%'  # Exibe a porcentagem com 2 casas decimais

**startangle = 90**

Startangle determina o ângulo de rotação inicial para o gráfico de pizza.

O valor de 90 faz com que a primeira fatia comece a partir do topo do gráfico (a linha horizontal que divide o gráfico em duas metades).

O valor de 0 faria o gráfico começar da parte direita (leste), enquanto 90 inicia no topo (norte), e 180 começaria da parte inferior (sul).

```python
# Exemplo

# Dados de exemplo: Participação de mercado das empresas
empresas = ['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D']
participacao = [40, 25, 20, 15]

# Criando o gráfico de pizza
plt.pie(participacao, labels=empresas, autopct='%1.1f%%', startangle=90)

# Adicionando título
plt.title('Participação de Mercado das Empresas')

# Exibindo o gráfico
plt.show()
```

## 5.5 Heatmap: Correlação entre as vendas de diferentes produtos

<code>**sns.heatmap(data_frame.corr(), annot = True, cmap = ' ')**</code>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Criando um DataFrame com vendas de produtos
vendas = pd.DataFrame({
    'Produto A': [100, 150, 120, 130, 140, 160, 180],
    'Produto B': [80, 90, 110, 120, 115, 130, 140],
    'Produto C': [60, 75, 90, 100, 105, 110, 120]
}, index=['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom'])

# Gerando o Heatmap
sns.heatmap(vendas.corr(), annot=True, cmap="coolwarm")

#vendas.corr() calcula a matriz de correlação dos dados fornecidos
#annot = True  adiciona os valores das correlações diretamente nas células do mapa de calor.
# cmap = '' define a paleta de cores (colormap) usada para representar os valores no mapa de calor

plt.title("Correlação entre Vendas de Produtos")
plt.show()
```

# 7. Salvando o gráfico em PNG ou PDF
Após criar o gráfico, você pode salvá-lo como uma imagem ou arquivo PDF:

## Matplotlib

Neste exemplo, vamos criar um gráfico de linha mostrando o crescimento de vendas ao longo dos meses e, em seguida, salvar o gráfico como uma imagem PNG e como um PDF.

```python
import matplotlib.pyplot as plt

# Dados de exemplo
meses = ['Jan', 'Feb', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [50, 80, 70, 90, 100, 110]

# Criando o gráfico
plt.plot(meses, vendas, marker='o', color='b', label='Vendas')

# Adicionando título e rótulos aos eixos
plt.title('Crescimento de Vendas')

# Salvando o gráfico como imagem PNG
plt.savefig('grafico_vendas.png')  # Salva como imagem PNG

# Salvando o gráfico como PDF
plt.savefig('grafico_vendas.pdf')  # Salva como PDF

# Exibindo o gráfico
plt.show()
```

## Seaborn
O Seaborn é construído sobre o Matplotlib, então as funções para salvar gráficos são as mesmas.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dados de exemplo
dados = sns.load_dataset('tips')

# Criando o gráfico de dispersão
sns.scatterplot(data=dados, x='total_bill', y='tip', hue='time')

# Adicionando título
plt.title('Gastos no Restaurante vs Gorjeta')

# Salvando o gráfico como imagem PNG
plt.savefig('grafico_seaborn.png')  # Salva como imagem PNG

# Salvando o gráfico como PDF
plt.savefig('grafico_seaborn.pdf')  # Salva como PDF

# Exibindo o gráfico
plt.show()
```
