### <font color=yellow> **1. Instalando e importando a biblioteca** </font>
O ```pandas``` é uma biblioteca muito utilizada para análise e manipulação de dados em Python. Com ele, é fácil trabalhar com dados tabulares, permitindo operações rápidas como seleção, filtragem e agregação. Além disso, ele facilita a leitura de dados de vários formatos e a preparação para análises e visualizações mais avançadas.

````python
# Instalando a biblioteca
# !pip install pandas

# Importando o pandas
import pandas as pd
````

### <font color=yellow> **2. Entendendo Series e DataFrames** </font>
Um objeto do tipo ```Series``` no pandas é uma matriz unidimensional que contém dados de qualquer tipo, incluindo inteiros, flutuantes, strings, booleanos, objetos python etc. Um ```DataFrame``` é composto de muitas séries que funcionam como colunas. Ambos são objetos tabulares.

````python
# Criando uma Series a partir de uma lista
temperaturas = pd.Series([30, 33, 40, 27, 18])
temperaturas
````

````python
# Alterando o índice de uma Series
temperaturas = pd.Series([30, 33, 40, 27, 18],
                         index = ["Seg", "Ter", "Qua", "Qui", "Sex"])
temperaturas
````

````python
# Acessando um valor da Series pelo índice
temperaturas["Qui"]
````

````python
# Criando um DataFrame a partir de um dicionário
pessoas = pd.DataFrame({'Nome': ['Maria', 'João', 'Ana'],
                      'Idade': [25, 30, 22],
                      'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']})
pessoas
````

````python
# Acessando colunas de um DataFrame com um par de colchete --> formato Series
pessoas['Nome']
````

````python
# Acessando colunas de um DataFrame com dois pares de colchetes --> formato DataFrame
pessoas[['Nome']]
````

````python
# Acessando intervalos de linhas com colchetes
# pessoas[:1]
pessoas[1:3]
````

### <font color=yellow> **3. Leitura, visualização e tratamento de bases de dados** </font>
O pandas nos fornece uma série de funcionalidades de leitura de dados. Aqui iremos utiliza o ```pd.read_csv()```, pois muitas bases de dados se encontram nesse formato. Mas, dê uma olhada na documentação oficial do pandas que você irá encontrar diversas formas de leitura de dados em diversos formatos: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

````python
# Importando um csv para um DataFrame
db_weather = pd.read_csv('weatherHistory.csv', sep=',')
db_weather
````

````python
# .head() para visualizar as primeiras linhas
db_weather.head(2)
````

````python
# .tail() para visualizar as últimas linhas
db_weather.tail(6)
````

````python
# Conferindo as dimensões do DataFrame a partir do '.shape'
db_weather.shape
````

````python
# Conferindo os tipos de dados no DataFrame a partir do '.dtypes'
db_weather.dtypes
````

````python
# Conferindo os tipos de dados no DataFrame a partir do '.dtypes'
db_weather.info()
````

````python
# Conferindo a quantidade de valores nulos no DataFrame a partir do '.isnull().sum()'
# db_weather.isnull()
db_weather.isnull().sum()
````

````python
# Conferindo a quantidade de valores ausentes no DataFrame a partir do '.isna().sum()'
# db_weather.isna()
db_weather.isna().sum()
````

````python
# Removendo valores nulos
db_weather.dropna()
````

````python
# Preenchendo valores nulos
db_weather.fillna(0).isna().sum()
````

````python
# Acessando as colunas de um DataFrame
db_weather.columns
````

````python
# Acessando o índice de um DataFrame
db_weather.index
````

````python
# Configurando uma coluna do DataFrame como index
db_weather = db_weather.set_index('Formatted Date')
db_weather
````

````python
# Desfazendo a configuração de uma coluna como index
db_weather = db_weather.reset_index()
db_weather.head(3)
````

````python
# Análise descritiva de um DataFrame
db_weather.describe()
````

````python
# Podemos usar .astype() para converter tipos de dados
db_weather[['Humidity', 'Visibility (km)']] = db_weather[['Humidity', 'Visibility (km)']].astype(str)
db_weather.dtypes
````

````python
# Renomendo colunas ou linhas
#db_weather = db_weather.rename(columns={'Temperature (C)': 'Temperatura (C)',
                                        #'Summary': 'Resumo'})
db_weather = db_weather.rename(index={0: 'Zero', 1: 'Um'})
db_weather.head(3)
````

````python
# Removendo colunas ou linhas
# db_weather = db_weather.drop(columns=['Humidity'])
db_weather = db_weather.drop(index=['Zero', 'Um'])
db_weather.head(3)
````

````python
# Utilizando máscaras para filtrar dados em DataFrames
# db_weather[db_weather['Temperatura (C)'] > 20]
# db_weather[db_weather['Resumo'] == 'Partly Cloudy']
db_weather[(db_weather['Temperatura (C)'] > 20) & (db_weather['Resumo'] == 'Partly Cloudy')]
````

### <font color=yellow> **4. Operações matemáticas com DataFrames** </font>
Para realizar análises e transformações de dados, é comum precisar de operações matemáticas com DataFrames. Em pandas, o uso de operações matemáticas é direto e flexível

````python
# Soma --> .sum()
# db_weather.sum()
db_weather[['Temperatura (C)', 'Pressure (millibars)']].sum()
````

````python
# Média --> .mean()
#db_weather.mean()
db_weather[['Temperatura (C)', 'Pressure (millibars)']].mean()
````

````python
# Desvio padrão --> .std()
# db_weather.std()
db_weather[['Temperatura (C)', 'Pressure (millibars)']].std()
````

````python
# Operações com valores constantes
db_weather['Temperatura (C)'] + 5
````

````python
# Operações entre colunas
db_weather['Temperatura (C)'] * db_weather['Pressure (millibars)']
````

````python
# Operações com funções
db_weather['Temperatura (C)'].sum() / len(db_weather['Pressure (millibars)'])
````
