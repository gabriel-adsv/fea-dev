### <font color=yellow> **2. Entendendo Series e DataFrames** </font>
Um objeto do tipo ```Series``` no pandas  uma matriz unidimensional que contm dados de qualquer tipo, incluindo inteiros, flutuantes, strings, booleanos, objetos python etc. Um ```DataFrame```  composto de muitas sries que funcionam como colunas. Ambos so objetos tabulares.

````python
# Criando uma Series a partir de uma lista
temperaturas = pd.Series([30, 33, 40, 27, 18])
temperaturas
````

````python
# Alterando o ndice de uma Series
temperaturas = pd.Series([30, 33, 40, 27, 18],
                         index = ["Seg", "Ter", "Qua", "Qui", "Sex"])
temperaturas
````

````python
# Acessando um valor da Series pelo ndice
temperaturas["Qui"]
````

````python
# Criando um DataFrame a partir de um dicionrio
pessoas = pd.DataFrame({'Nome': ['Maria', 'Joo', 'Ana'],
                      'Idade': [25, 30, 22],
                      'Cidade': ['So Paulo', 'Rio de Janeiro', 'Belo Horizonte']})
pessoas
````

````python
# Acessando colunas de um DataFrame com um par de colchete --> formato Series
pessoas['Nome']
````

````python
# Acessando colunas de um DataFrame com dois pares de colchetes --> formato DataFrame
pessoas[['Nome']]
````

````python
# Acessando intervalos de linhas com colchetes
# pessoas[:1]
pessoas[1:3]
````

### <font color=yellow> **3. Leitura, visualizao e tratamento de bases de dados** </font>
O pandas nos fornece uma srie de funcionalidades de leitura de dados. Aqui iremos utiliza o ```pd.read_csv()```, pois muitas bases de dados se encontram nesse formato. Mas, d uma olhada na documentao oficial do pandas que voc ir encontrar diversas formas de leitura de dados em diversos formatos: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

````python
# Importando um csv para um DataFrame
db_weather = pd.read_csv('weatherHistory.csv', sep=',')
db_weather
````

````python
# .head() para visualizar as primeiras linhas
db_weather.head(2)
````

````python
# .tail() para visualizar as ltimas linhas
db_weather.tail(6)
````

````python
# Conferindo as dimenses do DataFrame a partir do '.shape'
db_weather.shape
````

````python
# Conferindo os tipos de dados no DataFrame a partir do '.dtypes'
db_weather.dtypes
````

````python
# Conferindo os tipos de dados no DataFrame a partir do '.dtypes'
db_weather.info()
````

````python
# Conferindo a quantidade de valores nulos no DataFrame a partir do '.isnull().sum()'
# db_weather.isnull()
db_weather.isnull().sum()
````

````python
# Conferindo a quantidade de valores ausentes no DataFrame a partir do '.isna().sum()'
# db_weather.isna()
db_weather.isna().sum()
````

````python
# Removendo valores nulos
db_weather.dropna()
````

````python
# Preenchendo valores nulos
db_weather.fillna(0).isna().sum()
````

````python
# Acessando as colunas de um DataFrame
db_weather.columns
````

````python
# Acessando o ndice de um DataFrame
db_weather.index
````

````python
# Configurando uma coluna do DataFrame como index
db_weather = db_weather.set_index('Formatted Date')
db_weather
````

````python
# Desfazendo a configurao de uma coluna como index
db_weather = db_weather.reset_index()
db_weather.head(3)
````

````python
# Anlise descritiva de um DataFrame
db_weather.describe()
````

````python
# Podemos usar .astype() para converter tipos de dados
db_weather[['Humidity', 'Visibility (km)']] = db_weather[['Humidity', 'Visibility (km)']].astype(str)
db_weather.dtypes
````

````python
# Renomendo colunas ou linhas
#db_weather = db_weather.rename(columns={'Temperature (C)': 'Temperatura (C)',
                                        #'Summary': 'Resumo'})
db_weather = db_weather.rename(index={0: 'Zero', 1: 'Um'})
db_weather.head(3)
````

````python
# Removendo colunas ou linhas
# db_weather = db_weather.drop(columns=['Humidity'])
db_weather = db_weather.drop(index=['Zero', 'Um'])
db_weather.head(3)
````

````python
# Utilizando mscaras para filtrar dados em DataFrames
# db_weather[db_weather['Temperatura (C)'] > 20]
# db_weather[db_weather['Resumo'] == 'Partly Cloudy']
db_weather[(db_weather['Temperatura (C)'] > 20) & (db_weather['Resumo'] == 'Partly Cloudy')]
````

### <font color=yellow> **4. Operaes matemticas com DataFrames** </font>
Para realizar anlises e transformaes de dados,  comum precisar de operaes matemticas com DataFrames. Em pandas, o uso de operaes matemticas  direto e flexvel

````python
# Soma --> .sum()
# db_weather.sum()
db_weather[['Temperatura (C)', 'Pressure (millibars)']].sum()
````

````python
# Mdia --> .mean()
#db_weather.mean()
db_weather[['Temperatura (C)', 'Pressure (millibars)']].mean()
````

````python
# Desvio padro --> .std()
# db_weather.std()
db_weather[['Temperatura (C)', 'Pressure (millibars)']].std()
````

````python
# Operaes com valores constantes
db_weather['Temperatura (C)'] + 5
````

````python
# Operaes entre colunas
db_weather['Temperatura (C)'] * db_weather['Pressure (millibars)']
````

````python
# Operaes com funes
db_weather['Temperatura (C)'].sum() / len(db_weather['Pressure (millibars)'])
````
