# Diff e Pct Change

Diff: Calcula a subtração entre um elemento e o anterior na mesma coluna (ou linha) de um DataFrame.

Pct Change: Calcula a mesma variação do diff, porém, em formato percentual.

```python
import yfinance as yf

acao_apple = yf.Ticker("AAPL")
apple = acao_apple.history(period="1mo")
```

```python
apple.head()
```

```python
apple = apple.reset_index()
```

```python
apple = apple[["Date","Close"]]
```

```python
apple = apple.rename(columns={"Date":"data","Close":"preco_fechamento"})
```

```python
apple["data"] = apple["data"].astype(str)
apple["data"] = apple["data"].str[:10]
apple["data"] = pd.to_datetime(apple["data"],format="%Y-%m-%d")
```

```python
apple.head()
```

```python
apple["variacao_diaria_diff"] = apple["preco_fechamento"].diff()
```

```python
apple
```

```python
# diff e period: 

# O argumento period define a quantidade de linhas anteriores no cálculo do diff. 
# No primeiro exemplo, nós comparamos o valor da linha, com o valor da linha antecessor (ou seja, subimos uma linha 1)
# Com o argumento periods, podemos definir "quantas linhas subir" para vermos esse cálculo 


apple["variacao_diaria_diff2"] = apple["preco_fechamento"].diff(periods=3)
```

```python
apple.tail()
```

```python
# pct change 

# realiza o mesmo, porém em porcentagem

apple["variacao_diaria_percentual"] = apple["preco_fechamento"].pct_change()
```

```python
apple.tail()
```

```python
apple["variacao_diaria_percentual2"] = apple["preco_fechamento"].pct_change(periods=3)
```
