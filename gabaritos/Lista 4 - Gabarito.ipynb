Imagine que seu chefe pediu para voc, um pequeno gafanhoto (tambm conhecido como estagirio), calcular a quantidade de gua, caf e lanches necessria para um evento corporativo. Ele esperava que voc fizesse tudo manualmente com caneta e calculadora, tal qual os astecas, mas, como voc est aprendendo Python, decidiu aplicar os seus conhecimentos e ganhar tempo.

a) Crie uma funo chamada ```calcular_suprimentos()``` que receba como argumento obrigatrio o nmero de participantes, calcule a quantidade necessria de cada item e imprima ao final um relatrio que ser enviado ao time de *Suprimentos*. Para isso, considere que:
- Cada pessoa, por padro, consome 2 garrafas de gua, 2 xcaras de caf e 3 unidades de lanche.
- Como alguns eventos podem durar mais ou menos, sua funo deve incluir argumentos opcionais para ajustar essas quantidades conforme necessrio: ```agua_por_pessoa```, ```cafe_por_pessoa``` e ```lanche_por_pessoa```.

Ao final, faa um exemplo de uso da funo, preenchendo tambm os argumentos opcionais da funo com valores diferentes do padro.

````python
##### SEU CODIGO COMEA AQUI #####

# Funo para calcular a quantidade de suprimentos necessrios
def calcular_suprimentos(quantidade_participantes, agua_por_pessoa=2, cafe_por_pessoa=2, lanche_por_pessoa=3):

    # Calcula a quantidade total de cada item
    total_agua = quantidade_participantes * agua_por_pessoa
    total_cafe = quantidade_participantes * cafe_por_pessoa
    total_lanche = quantidade_participantes * lanche_por_pessoa

    # Retorna o resultado em um dicionrio
    print (
        f"Relatrio de Suprimentos para o Evento\n"
        f"{'-'*40}\n"
        f"Quantidade de participantes: {quantidade_participantes}\n"
        f"Suprimentos necessrios:\n"
        f"- gua: {total_agua} ml\n"
        f"- Caf: {total_cafe} xcaras\n"
        f"- Lanches: {total_lanche} unidades"
    )

# Exemplos de uso
calcular_suprimentos(quantidade_participantes=15, agua_por_pessoa=4, cafe_por_pessoa=3)  # Usando valores personalizados
````

b) Agora, atualize a funo para que ela permita receber outros itens personalizados conforme necessrio  por exemplo, ```suco_por_pessoa``` ou ```doces_por_pessoa```. A funo deve calcular a quantidade total de cada item, incluindo os itens adicionais passados. Imprima ao final um relatrio com as quantidades necessrias de cada item, incluindo os itens extras que foram especificados.

````python
##### SEU CODIGO COMEA AQUI #####

# Funo para calcular a quantidade de suprimentos necessrios
def calcular_suprimentos(quantidade_participantes, agua_por_pessoa=2, cafe_por_pessoa=2, lanche_por_pessoa=3, **outros):

    # Calcula a quantidade total de cada item padro
    total_agua = quantidade_participantes * agua_por_pessoa
    total_cafe = quantidade_participantes * cafe_por_pessoa
    total_lanche = quantidade_participantes * lanche_por_pessoa

    # Inicializa o dicionrio com os itens padro
    suprimentos = {
        'gua': total_agua,
        'cafe': total_cafe,
        'lanche': total_lanche
    }

    # Calcula os itens adicionais passados por **kwargs
    for item, quantidade_por_pessoa in outros.items():
        suprimentos[item] = quantidade_participantes * quantidade_por_pessoa

    # Gera o relatrio em formato de texto
    print (
        f"Relatrio de Suprimentos para o Evento\n"
        f"{'-'*40}\n\n"
        f"Quantidade de participantes: {quantidade_participantes}\n\n"
        f"Suprimentos necessrios:\n"
    )

    for item, total in suprimentos.items():
        print(f"- {item.capitalize()}: {total} unidades\n")

# Exemplo de uso
calcular_suprimentos(20, agua_por_pessoa=2, cafe_por_pessoa=3, suco_por_pessoa=1, doces_por_pessoa=2)
````

## **Exerccio 3**: Descontos da Black Friday!

3.png

Imagine que voc trabalha no setor de vendas de uma empresa e vocs esto se preparando para o famoso dia da **Black Friday**, precisando calcular os preos com desconto para uma lista de preos de produtos.

a) Crie uma funo ```calcular_precos_com_descontos()``` que, usando ```map``` e ```lambda```, aplique um percentual de 10% de desconto a cada valor em uma lista de preos. A funo deve retornar uma nova lista com os preos j ajustados.

````python
# Definindo os preos e descontos
precos = [100, 150, 200, 250, 300]
desconto = 0.10

##### SEU CODIGO COMEA AQUI #####

# Funo para calcular os descontos
def calcular_precos_com_descontos(precos, desconto):

    # Usando map e lambda para aplicar o desconto a cada preo
    precos_com_desconto = map(lambda preco: round(preco * (1 - desconto), 2), precos)

    # Convertendo o resultado de map para uma lista
    return list(precos_com_desconto)

# Utilizando a funo
precos_com_desconto = calcular_precos_com_descontos(precos, desconto)
print(precos_com_desconto)
````

b) Modifique a funo ```calcular_precos_com_descontos()``` para que ela permita aplicar diferentes percentuais de desconto para cada item respectivamente.

````python
precos = [100, 150, 200, 250, 300]
descontos = [0.10, 0.15, 0.20, 0.25, 0.30]

##### SEU CODIGO COMEA AQUI #####

# Funo para calcular os descontos com percentuais variveis
def calcular_precos_com_descontos(precos, descontos):
    precos_com_desconto = map(lambda preco, desconto: round(preco * (1 - desconto), 2), precos, descontos)
    return list(precos_com_desconto)

# Exemplo de uso com diferentes descontos para cada produto
precos_com_desconto = calcular_precos_com_descontos(precos, descontos)
print(precos_com_desconto)
````

## **Exerccio 4**: Inventrio da Vila do Chaves

Voc foi contratado pelo Seu Barriga para organizar um sistema organizado de inventrio com os itens da Vila do Chaves. Para isso, crie uma funo ```atualizar_inventario()``` que receba um dicionrio de inventrio e oferea as seguintes opes:

- Consultar o inventrio;
- Adicionar ou substituir um item e a sua quantidade no inventrio;
- Verificar o total de itens presentes no inventrio.

Ao final do exerccio, d exemplos de uso da funo para cada uma das operaes criadas.

````python
# Inventrio inicial da Vila
inventario_vila = {"Barril do Chaves": 1,
                   "Vestido da Chiquinha": 1,
                   "Chapu do Seu Madruga": 1,
                   "Pirulito do Quico": 5,
                   "Buqu de flores da dona Florinda": 7}

##### SEU CODIGO COMEA AQUI #####

# Funo para atualizar o inventrio da Vila do Chaves
def atualizar_inventario(inventario, operacao, item=None, quantidade=0):

    # Operao: Consultar itens do personagem
    if operacao == "consultar":
        return f"Inventrio: {inventario}"

    # Operao: Adicionar item ao inventrio do personagem
    elif operacao == "adicionar":
        inventario[item] = quantidade
        return f"Item '{item}' adicionado ao inventrio."

    # Operao: Verificar o total de itens do personagem
    elif operacao == "total":
        total = sum(inventario.values())
        return f"O inventrio possui um total de {total} itens."

    # Se a operao no for reconhecida
    else:
        return "Operao invlida. Escolha entre 'consultar', 'adicionar', 'remover', ou 'total'."

# Exemplos de uso
# Adicionando itens ao inventrio do Chaves
print(atualizar_inventario(inventario_vila, "adicionar", item="bola de futebol", quantidade=2))

# Consultando o inventrio atualizado do Chaves
print(atualizar_inventario(inventario_vila, "consultar"))

# Removendo um item do inventrio do Seu Madruga
print(atualizar_inventario(inventario_vila, "remover", item="chapu"))

# Verificando o total de itens do Quico
print(atualizar_inventario(inventario_vila, "total"))
````

## **Exerccio Desafio**: Removendo Outliers

boxplot.png

Imagine que voc e seus amigos resolveram ver quem mais maratonou sries na Netflix na ltima semana. Cada um anotou o nmero de horas que passou assistindo, mas tem sempre aquele amigo que foi longe demais. Esse nmero est muito fora da curva e distorce a mdia da galera. Ento, sua tarefa  criar uma funo chamada ```remover_outliers()``` que receba uma lista com todas as horas assistidas e remova os valores extremos usando o mtodo do intervalo interquartlico (IQR) para depois calcular a mdia com os valores ajustados. Para isso:

1. Organize os dados em ordem crescente.

2. Crie uma funo que encontre o primeiro quartil (Q1) e o terceiro quartil (Q3):
    - **Q1**  o valor que marca os 25% mais baixos dos dados, ou seja, 25% dos dados esto abaixo desse valor.
    - **Q3**  o valor que marca os 75% dos dados, o ponto onde ficam os 25% mais altos dos valores.

3. Nessa mesma funo, calcule o **IQR**, ou seja, a diferena entre o terceiro quartil e o primeiro quartil.

4. Agora, utilizando a funo anteior, crie uma funo que determine e exclua os outliers, de modo que:
    - O limite inferior seja definido como *Q1  1.5 * IQR*.
    - O limite superior seja definido como *Q3 + 1.5 * IQR*.

````python
# Lista de horas assistidas na Netflix
horas_netflix = [4, 20, 6, 7, 30, 8, 10, 12, 15, 7000, 18, 100]

##### SEU CODIGO COMEA AQUI #####

# Funo para calcular Q1 e Q3
def calcular_quartis(dados):

    # Ordenando os dados
    dados_ordenados = sorted(dados)
    n = len(dados_ordenados)

    # Calculando Q1 (25 percentil)
    q1 = dados_ordenados[int(n * 0.25)]

    # Calculando Q3 (75 percentil)
    q3 = dados_ordenados[int(n * 0.75)]

    # Calculando o IQR
    iqr = q3 - q1

    return q1, q3, iqr

# Funo para remover outliers
def remover_outliers(dados):

    # Calculando Q1, Q3 e IQR
    q1, q3, iqr = calcular_quartis(dados)

    # Calculando os limites
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr

    # Filtrando dados dentro dos limites
    dados_sem_outliers = []

    for i in dados:
        if (i >= limite_inferior) and (i <= limite_superior):
            dados_sem_outliers.append(i)

    return dados_sem_outliers

# Removendo outliers e calculando a mdia
novos_dados = remover_outliers(horas_netflix)
print(f"Dados sem outliers: {novos_dados}")
media = sum(novos_dados) / len(novos_dados)
print(f"Mdia de horas assistidas (sem outliers): {round(media,2)}h")
````
